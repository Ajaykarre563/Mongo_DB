# **$in Query:**

# **db.users.find({ address: { $in: ["USA", "UK"] } })**



```
db.users.find({ address: { $in: ["USA", "UK"] } })

```



# ** Explanation:**
db.users.find(...): Queries the users collection.

{ address: { $in: [...] } }: Checks if any value in the address array matches one of the values in ["USA", "UK"].

Since in your documents address is an array like ["New York", "USA"], this will match if "USA" or "UK" is present anywhere inside that array.



# **$nin Query:**



```
db.users.find({ address: { $nin: ["USA", "UK"] } })

```

# **$all**

# ** Meaning of $all:**
This query returns documents where the address array contains both "London" and "UK"  in any order.




```
db.users.find({ address: { $all: ["London", "UK"] } })

```
