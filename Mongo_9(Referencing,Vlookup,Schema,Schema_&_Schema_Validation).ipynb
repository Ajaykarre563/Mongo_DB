# **📘 What lookup Does in MongoDB**
In MongoDB, the $lookup stage is used in the aggregation pipeline to perform a left outer join between two collections.

# **🔄 How It Works**
$lookup looks up a value from the current document in one collection and tries to match it with documents in another collection.

When a match is found, the matching document(s) are embedded as an array in the output.



image.png

image.png



```
db.orders.aggregate([
    // Stage 1: Lookup to join with users collection
    {
        $lookup: {
            from: "users",
            localField: "customer_id",
            foreignField: "_id",
            as: "customer_info"
        }
    }
])
```

# **📘 Defining Schema & Schema Validation in MongoDB**
MongoDB is schema-less by default, but you can define a schema with validation rules using JSON Schema. This helps ensure data consistency by enforcing field types and structure.



# **1. Defining a Schema with Validation**
You can define a schema using db.createCollection() with a validator using $jsonSchema.



```
db.createCollection('users', {
  validator: {
    $jsonSchema: {
      bsonType: 'object',
      required: ['name', 'email', 'address'],
      properties: {
        name: {
          bsonType: 'string',
          description: 'Name is a required text field.'
        },
        email: {
          bsonType: 'string',
          description: 'Email is a required text field.'
        },
        address: {
          bsonType: 'object',
          required: ['street', 'city', 'country'],
          properties: {
            street: { bsonType: 'string' },
            city: { bsonType: 'string' },
            country: { bsonType: 'string' }
          }
        },
        gender: {
          bsonType: 'string'
        },
        age: {
          bsonType: 'int',
          minimum: 0
        }
      }
    }
  }
});

```
