**In MongoDB, a partial index is an index that only includes the documents in a collection that meet a specified filter expression. This is useful for:**

Reducing index size

Improving write performance

Optimizing queries that target a subset of data





```
db.users.createIndex(
    { Age: 1 },
    { partialFilterExpression: { Gender: 'Male' } }
)
```
