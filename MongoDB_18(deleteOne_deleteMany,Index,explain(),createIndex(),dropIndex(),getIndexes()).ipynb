# **Understanding Index in MongoDB**
In MongoDB, an index is a data structure that improves the speed of search queries on a collection.



# **🔍 Why Use Indexes?**
Without indexes, MongoDB must scan every document in a collection — this is slow for large datasets.

Indexes let MongoDB quickly find matching documents by pointing directly to their location, just like an index in a book.
