# **Key Points:**

Ordered Insertion with insertMany() means that each document is inserted one after the other in sequence.

If any document fails to insert (e.g., due to validation error), the process stops at that point.

Already inserted documents remain in the database; the system does not roll back the previous successful inserts.

# **🔹 What is Write Concern?**
Write Concern defines the level of acknowledgment the database must provide for a write operation (like insert, update, delete) to be considered successful.

# **✅ Key Points:**
It controls the durability and replication guarantee of a write.

It determines how many nodes must confirm the write, and whether the data must be written to disk.

Used to balance performance and reliability.

# **📘 What Is the Journal?**
The journal is a write-ahead log that records changes to the database before they are applied to the actual data files. This ensures that, in the event of a crash (such as power failure or server crash), MongoDB can recover to a consistent state by replaying the journal entries.

# **🔧 How It Works:**
Write Operation Initiated
A client issues a write operation (insert, update, or delete).

**Data Written to Journal**

MongoDB writes the change to the journal file before writing it to the main data files on disk.

**Journal Flushed to Disk**

The journal is flushed to disk every 100 milliseconds by default (configurable).

**Durability Ensured**

Even if the server crashes after the journal is written but before the actual data file is updated, MongoDB can use the journal to recover.

**MongoDB does NOT roll back the previous successful inserts if one insert fails by default.**

If the first document is inserted successfully, and the second one fails (e.g., duplicate _id), then:

The first stays inserted

The failed ones are not inserted

There is no rollback of already inserted documents
