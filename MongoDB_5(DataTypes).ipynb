# **String:**

A string is a text value enclosed in quotes (either single ' or double ").

Strings are used to represent words, sentences, or any sequence of characters.



```
{
  "name": "Alice Johnson",
  "address": "123 Main St"
}

```

# **Example:**



```
db.users.insertOne({
  name: "Jone Doe",                 // String: User's name
  age: 28,                          // Number: User's age
  isAdmin: true,                   // Boolean: Is the user an admin?
  dob: new Date(),                 // Date object: Current date/time (user's date of birth here)
  createAt: new Timestamp(),       // Timestamp: Represents creation time
  address: null,                   // Null: No address provided
  gender: undefined                // Undefined: Value not set
})

```
