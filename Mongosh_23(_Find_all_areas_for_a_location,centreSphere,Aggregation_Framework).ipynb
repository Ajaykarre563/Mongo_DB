Complex queries like grouping, filtering, sorting, or joining data.

Data transformation (e.g., calculating averages, totals).

An alternative to writing complex JavaScript logic in your app.

# **FirstName And LastName first is upper:**



```
db.employees.aggregate([
  {
    $project: {
      _id: 0,
      name: {
        $concat: [
          { $toUpper: { $substrCP: ["$firstname", 0, 1] } },   // First letter uppercase
          { $substrCP: ["$firstname", 1, { $subtract: [ { $strLenCP: "$firstname" }, 1 ] }] }, // Rest of first name
          " ",
          { $toUpper: { $substrCP: ["$lastname", 0, 1] } },    // First letter uppercase
          { $substrCP: ["$lastname", 1, { $subtract: [ { $strLenCP: "$lastname" }, 1 ] }] }  // Rest of last name
        ]
      },
      gender: 1,
      email: 1
    }
  }
])

```
