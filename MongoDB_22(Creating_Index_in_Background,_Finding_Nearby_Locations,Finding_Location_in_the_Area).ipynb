# **🌍 What is Geospatial Data in MongoDB?**
Geospatial data refers to information that describes locations on the Earth's surface — like coordinates, shapes, or areas.

**In MongoDB, geospatial data is stored in specific formats that allow you to**:

Store points, lines, and polygons

**Perform geospatial queries, like:**

“Find places near this location”

“Check if a point is inside a certain area”

“Sort places by distance from a location”

# **finding places inside a polygon area using MongoDB and GeoJSON, based on the points you shared (p1 to p4):**



```
// Define polygon points (longitude, latitude)
const p1 = [77.54868, 13.01133];
const p2 = [77.5552, 13.0114];
const p3 = [77.5549, 13.00839];
const p4 = [77.54795, 13.00868];

// Query to find places inside the polygon
db.places.find({
  location: {
    $geoWithin: {
      $geometry: {
        type: "Polygon",
        coordinates: [[
          p1, p2, p3, p4, p1  // Closed polygon loop (first = last)
        ]]
      }
    }
  }
});

```
