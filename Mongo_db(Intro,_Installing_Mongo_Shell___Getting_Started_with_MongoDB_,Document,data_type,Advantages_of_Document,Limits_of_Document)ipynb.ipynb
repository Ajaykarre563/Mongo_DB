# **What is MongoDB?**

MongoDB is an open-source, document-oriented NoSQL database. Unlike traditional relational databases that store data in tables and rows, MongoDB stores data in flexible, JSON-like documents. This makes it ideal for handling complex, unstructured, and rapidly evolving data.

# **NOTE: NoSQL stands for Not Only SQL**

# **1. Relational Database (SQL):**
Data is stored in tables (rows and columns).

Example shown is a Customers table with structured fields like ID, Name, Gender, etc.

# **MongoDB DATABASE**
 MongoDB databases stores data in a collections. Each collection is a collection of documents, which stores data in the form of key-value pair. This key-value pair is called as field.

 In MongoDB database, there is no pre-defined schema. Two or more documents can have different fields & different number of fields. We can also define a schema on the collection if required.

 In MongoDB, we can store related data in the same document. This makes querying related data faster. But it might also lead to redundant data.

# **MongoDB Documents**
 A document of a collection stores data in a JSON-like format. This format is called as BSON.

 A MongoDB collection is schema-less. We can store two different types of data with different fields & different number of fields, in the same collection.

 We can also store related data of a document in the same document. This makes querying related data faster.

 Each key-value pair in a document is called as field of the document. Each field represents an information about the data entity that document is representing.

# **What is a Document?**

MongoDB is a NoSQL database that stores data in a flexible, JSON-like format called documents. Documents are the fundamental units of data in MongoDB, analogous to rows in relational databases, but far more flexible and expressive.

 A document is a data structure composed of field-value pairs, similar to a JSON object.

 It is stored in BSON (Binary JSON) format internally, which allows for faster processing and additional data types not supported by JSON (like Date, Decimal128, etc.).

# **Data Types:**

The image lists some common data types in MongoDB, which is a NoSQL database that stores data in BSON (Binary JSON) format. Here's a brief explanation of each data type shown:

1. **Integer**: Stores whole numbers, either 32-bit or 64-bit depending on the system. Used for numerical values like counts or IDs (e.g., `42`).

2. **String**: Stores sequences of characters, typically text. Its UTF-8 encoded in MongoDB (e.g., `"Hello, MongoDB"`).

3. **Boolean**: Stores `true` or `false` values. Useful for flags or binary states (e.g., `true`).

4. **Array**: Stores lists of values, which can be of mixed types. Arrays are ordered (e.g., `[1, "two", 3.0]`).

5. **Object**: Stores nested documents, essentially key-value pairs, allowing for hierarchical data (e.g., `{ "name": "Alice", "age": 25 }`).

6. **Null**: Represents a null or missing value (e.g., `null`).

7. **Date**: Stores dates and times, typically in UTC format. Useful for timestamps (e.g., `new Date("2025-05-12")`).

8. **ObjectId**: A special 12-byte type used as a default unique identifier for documents in MongoDB (e.g., `ObjectId("507f1f77bcf86cd799439011")`).

9. **Binary Data**: Stores binary data, like images or files, using a `BinData` type (e.g., for storing a PDF file).

10. **Regex**: Stores regular expressions for pattern matching in queries (e.g., `/^abc/`).

11. **Double**: Stores floating-point numbers for decimal values (e.g., `3.14`).

12. **etc.**: Indicates there are more types, like:
    - **Decimal128**: For high-precision decimal numbers (e.g., financial calculations).
    - **Timestamp**: For internal MongoDB use, often in replication.
    - **Code**: Stores JavaScript code (rarely used).

MongoDBs flexible schema allows these types to be mixed within documents, making it easy to store complex, unstructured data. Let me know if you'd like a deeper dive into any specific type!

# **Advantages of Document?**

**Flexibility:** Fields can differ across documents in the same collection.

**Embedded Relationships:** Allows storing related data together in one document (e.g., user data and their addresses).

**Rich Queries:** MongoDB provides powerful query operators to work with document structures (e.g., querying nested fields, arrays).

**Ease of Use:** JSON-like syntax is intuitive for developers familiar with JavaScript and JSON.

# **Limits of Document?**

**Size Limit:** A single document cannot exceed 16 MB in size. If larger, you should use GridFS for storing large files.

**Schema Management:** Lack of enforced schema can lead to inconsistent data if not carefully managed.
