# **✅ Covered Queries in MongoDB**
A covered query is a type of query in MongoDB where:

All the fields used in the query and in the returned result (projection) are covered by an index, and

The query does not need to scan the documents in the collection — it can get everything from the index.

# **🔍 Why Use Covered Queries?**
Faster performance because they avoid reading the full document.

Reduced disk I/O and improved efficiency.

# **🧠 Requirements for a Covered Query**
For a query to be covered:

An index must include all the fields in the query and projection.

The _id field must be excluded if it's not in the index (it's included by default in projections).
