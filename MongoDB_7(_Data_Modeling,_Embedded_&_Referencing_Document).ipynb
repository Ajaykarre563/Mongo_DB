# **what is mean by  Data Modeling in MongoDB?**

Designing the structure of your data to efficiently store and retrieve information. Unlike traditional relational databases, MongoDB is a NoSQL database that uses a flexible schema, allowing you to adapt to changing data requirements.

# **Key Takeaways:**
Data Modeling defines how data is organized and stored.

MongoDB uses NoSQL, not traditional relational models.

It supports schema flexibility, meaning documents in the same collection can have different fields.

It helps in adapting quickly to changes in data needs.

# **Advantages of Embedded Documents:**

**Improved Performance:** All related data (e.g., customer and orders) is in a single document, so you don’t need to query multiple collections. This makes queries faster.
# **Disadvantages of Embedded Documents:**

**Redundancy:** Can lead to duplicate data in the collection (e.g., "PlayStation 6" details repeated across documents).

**Higher Bandwidth:** Fetching a document with all embedded data requires more network bandwidth.

**Size Limit:** MongoDB documents cannot exceed 16 MB. Embedding large amounts of data might cause the document size to exceed this limit.
