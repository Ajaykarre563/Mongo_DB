```
**1. Query Selector**

**Used to filter documents in a collection.**

Comparison ($eq, $gt, $lt, etc.)

Logical ($and, $or, $not, etc.)

Element ($exists, $type, etc.)

Evaluation ($expr, $mod, etc.)

Array ($all, $size, $elemMatch, etc.)

Comments ($comment)

Geospatial ($geoWithin, $near, etc.)

**2. Projection**

**Used to select or format fields returned in a query.**

$  Project specific fields

$elemMatch  Project first matching array element

$meta  Metadata projection (e.g., textScore)

$slice  Return a subset of array elements
```
