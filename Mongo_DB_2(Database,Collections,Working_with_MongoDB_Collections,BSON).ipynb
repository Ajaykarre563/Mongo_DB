# **Third way to create the Collections:**

# **capped: true:**

This makes the collection a capped collection. A capped collection has a fixed size and automatically overwrites old documents when it reaches its limit, acting like a circular buffer.
Capped collections are useful for logs or caches where you only need recent data (e.g., the newest 100 documents).
Once set to true, the collection cannot be converted back to a non-capped collection.
# **size: 10000:**
Specifies the maximum size of the capped collection in bytes. Here, 10000 bytes is roughly 10 KB.
This limits how much disk space the collection can use. When the size limit is reached, MongoDB starts overwriting the oldest documents.
# **max: 100:**
Sets the maximum number of documents the capped collection can hold, in this case, 100 documents.
If the max limit is reached before the size limit, MongoDB will overwrite the oldest documents to make room for new ones.
Both size and max work together: the collection stops growing when either limit is hit.
# **autoIndexId: true:**
Ensures that MongoDB automatically creates an index on the _id field for each document in the collection.
By default, MongoDB creates this index for regular collections, but for capped collections, this is not automatic unless explicitly set to true.
Indexing the _id field improves query performance when searching by _id.

# **What is BSON?**

BSON (Binary JSON) is a binary-encoded serialization format for data interchange, just like JSON. We can say, BSON is JSON data in binary format.

It’s designed to be more efficient than JSON for storing and transmitting data, particularly in large-scale applications.
BSON is primarily used by MongoDB, but it can also be used in other applications that require a compact and efficient data format.
BSON supports more data types than JSON for storing the data efficiently.
