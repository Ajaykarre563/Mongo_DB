# **Adding multiple Documents:**



```
db.employees.updateOne(
  { name: "Jane Smith" },
  {
    $push: {
      skills: {
        $each: [
          { name: "backend", level: 4 },
          { name: "fullstack", level: 2 }
        ],
        $sort: { level: 1 },  // Sorts by level ascending
        $slice: 1             // Keeps only the top (lowest level) skill
      }
    }
  }
);

```



image.png

# **$addToSet in MongoDB is great for avoiding duplicates.**



```
db.employees.updateOne(
  { name: "Jane Smith" },
  {
    $addToSet: {
      skills: {
        name: "fullstack",
        level: 2
      }
    }
  }
);

```



image.png

image.png



```
db.employees.updateOne(
  { name: "Jane Smith" },
  {
    $pull: {
      skills: { name: "fullstack" }
    }
  }
);

```



image.png

# **In MongoDB, the $pop operator is used to remove an element from either end of an array:**

$pop: { field: 1 } → Removes the last element of the array.

$pop: { field: -1 } → Removes the first element of the array.

image.png



```
db.employees.updateOne(
   { name: "Jane Smith" },
   { $pop: { skills: 1 } }
)

```
