# **What is a Cursor in MongoDB?**

MongoDB doesn’t immediately return all documents. Instead, it returns a cursor — a pointer to the result set on the server.

image.png

image.png

image.png

image.png

# **Key Benefits:**

**Saves Network Bandwidth:** Instead of retrieving the entire result set in one go (which could be large), a cursor fetches data incrementally. This prevents overloading the network with a large amount of data.

**Improves Application Performance:** Since data is fetched in smaller chunks, the application processes smaller amounts at a time, leading to better performance.

**Prevents Huge Data Processing:** By fetching data incrementally, a cursor helps avoid overwhelming the application with too much data at once, reducing the risk of crashes.
