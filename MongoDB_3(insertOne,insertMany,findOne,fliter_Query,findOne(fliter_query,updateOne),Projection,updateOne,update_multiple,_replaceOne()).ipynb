# **What is mean by insertMany()?**

**insertMany() is used to insert multiple documents at once into the users collection.**

**Each document inside the array {} is a JavaScript object representing a user.**

**Fields like name, age, and gender are keys in these objects.**



```
db.users.insertMany([
  { name: 'sarah', gender: 'female' },
  { name: 'steve', age: 30 },
  { name: 'john', age: 54, gender: 'unknown' }
]);

```

# **✅ What is Projection?**
Projection controls which fields you want to include or exclude in your query result.

image.png

## **✅ Notes:**
Use 1 or true to include a field.

Use 0 or false to exclude a field.

You cannot mix include and exclude (except _id can be excluded separately).
